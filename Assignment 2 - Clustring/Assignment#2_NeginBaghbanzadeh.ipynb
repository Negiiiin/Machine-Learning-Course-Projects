### Q1 - Part a

```python
sum_squared_dists = []
number_of_clusters = range(1,10)
for k in number_of_clusters:
    kmeanModel = KMeans(n_clusters=k)
    kmeanModel.fit(X)
    sum_squared_dists.append(kmeanModel.inertia_)
```

```python
plt.figure(figsize=(16,8))
plt.plot(number_of_clusters, sum_squared_dists, 'bx-')
plt.xlabel('Number of Clusters')
plt.ylabel('Sum of Squared Distances')
plt.title('The Elbow Method showing the optimal Number of Clusters')
plt.show()
```

Based on the elbow graph above, having 6 clusters, doesn't make the much diffrence in making sum of squared distaances less than 7 clusters (making the model better). So a convenient number of clusters is **4** clusters.

```python
from sklearn import metrics
silhouette_score = []
number_of_clusters = range(2,10)
for k in number_of_clusters:
    kmeans_model_one_client = KMeans(n_clusters=k).fit(X)
    labels = kmeans_model_one_client.labels_
    silhouette_score.append(metrics.silhouette_score(X, labels, metric='euclidean'))
```

```python
plt.figure(figsize=(16,8))
plt.plot(number_of_clusters, silhouette_score, 'bx-')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.title('Silhoutte coefficient score per Cluster ID')
plt.show()
```

The Silhoutte coefficient score confirm k=5 too.

#### The following cell generates a list of all the days in the two years 2013-2014, which is helpful to answer part (b).  The cells below are only to give you a headstart.  You may or may not use these and come up with your own interpretation.


```python
from datetime import date, timedelta

d1 = date(2013, 1, 1)  # start date
d2 = date(2014, 12, 31)  # end date
delta = d2 - d1         # timedelta
daysyear = []
D = {0:'mon', 1:'tue', 2:'wed', 3:'thu', 4:'fri', 5:'sat', 6:'sun'}
for i in range(delta.days + 1):
    daysyear.extend([D[(d1 + timedelta(days=i)).weekday()]+"-"+str(d1 + timedelta(days=i))])
```

Since we have 2 clusters, one easy guess is that the clussters represent weekadys and weekends. So we're going to plot days of the year vs cluster ID to see if our alternitive hypothesis is right or not.

```python
X_one_client_df = pd.DataFrame(X_one_client)
X_one_client_df['daysyear'] = daysyear #[i[0:3] for i in daysyear]
optimal_k=2
optimal_model_one_client = KMeans(n_clusters=optimal_k).fit(X_one_client)
X_one_client_df['cluster'] = optimal_model_one_client.labels_
```

```python
plt.figure(figsize=(500,10))
plt.xticks(rotation=90)
plt.plot(X_one_client_df['daysyear'], X_one_client_df['cluster']==1)
plt.title("Days of Year VS. Cluster ID")
plt.xlabel("Days in Year")
plt.ylabel("Cluster ID")
```

If you look at the above graph(double click on it) you can see that most of the times, on weekends, the cluster ID is 0 and on weekdays it is 1. So we can say our alternitive hypothesis is right and the clusters represent week days and weekends.
